```python
#영어 명사의 부정관사를 결정하는 프로그램
```

```python
noun = input('명사를 입력하세요.')
if noun.startswith(tuple('aeiou')):
  particle = 'an'
else:
  particle = 'a'
print('{} {}'.format(particle, noun) )
```

```python
def get_particle(noun):
  '''영어 단어에 맞는 부정관사 변환'''
  if noun.startswith(tuple('aeiou')):
    particle = 'an'
  else:
    particle = 'a'
    return particle
```
