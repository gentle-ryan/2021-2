```python
today = input('오늘의 요일을 입력하세요: ')
N = int(input('며칠 후의 요일을 계산할까요? '))
week = '월화수목금토일'
for char in week:
  if char == today:
    n = week.index(today)
  else:
    pass
M = n+N
m = M%7
day = week[m]

print(f'오늘은 {today}요일이고, {N}일 후는 {day}요일입니다.')
```

```python
sentence = input('문장을 입력하세요: ')
sentence = sentence.lower()
words = ''
for char in sentence:
  if char.isalpha():
    words = words+char
  elif char == ' ':
    words = words + char
  else:
    pass

print(words)

words = words.split()
freq = {}

for word in words:
  if word in freq:
    freq[word] +=1
  else:
    freq[word] = 1

print(freq) 
```
