```python
S = 0 # 덧셈에 대한 항등원
i = 0 #1로 시작할 수 있도록
while S < 100:
  i = i + 1
  S = S + i
#반복이 끝난 뒤  

print(S, i)
```

```python
S = 0
for i in range(1, 11):
  S = S + i
  
print(S, i)
```

```python
word = 'egg'
for i, char in enumerate(word):
  print('the {}-th letter of "egg" is {}'.format(i+1, char))
```
