```python
#문장 내에서 특정 문자의 개수를 세는 프로그램
sentence = input('문장을 입력하세요: ')
char = input('찾고 싶은 문자르 입력하세요: ')
n = sentence.count(char)
print('{}의 출현 횟수 : {}'.format(char, n))
```

```python
# 문장 내 단어를 가나다순으로 정렬하는 프로그램
sentence = input('문장을 입력하세요: ')
words = sentence.split()
words.sort()
print(words)
```
